### Finding number of married men

We can create a set of conditions and pass it to numpy `where()` method to map a specific value to the result of the condition.

```python
conditions = [(trainDataFrame['sex']==0) & (trainDataFrame['marital-status']=='Married-civ-spouse')]
marriedMenNum = (numpy.where(conditions, 1, 0) == 1).sum()
print('\nNumber of married men :', marriedMenNum)
```

In this part we create a `conditons` set based of our need and then use `where` to map results to it.   
In the above code if condtions for a row is true, result would be 1, else it would be 0.   
We later use `sum()` to find number of 1's (number of rows in which condition was true).

## Part 9
### Generating plots   

In this part we use `hist` method to create distribution diagram for data of each column.   
However before that we create a function named `getColumnNameByIndex` to get the name of each column by giving an index number as input.

```python
def getColumnNameByIndex(index) :
    if(index==0) :
        return 'age'
    elif(index==1) :
        return 'workclass'
    elif(index==2) :
        return 'fnlwgt'
    elif(index==3) :
        return 'education'
    elif(index==4) :
        return 'education-num'
    elif(index==5) :
        return 'marital-status'
    elif(index==6) :
        return 'occupation'
    elif(index==7) :
        return 'relationship'
    elif(index==8) :
        return 'race'
    elif(index==9) :
        return 'sex'
    elif(index==10) :
        return 'capital-gain'
    elif(index==11) :
        return 'capital-loss'
    elif(index==12) :
        return 'hours-per-week'
    elif(index==13) :
        return 'native-country'
    elif(index==14) :
        return 'salary'
    else :
        return 'index-error'
```

An acceptable index number ranges between 0 to 14 as we have 15 columns (Including previously deleted ones).   
Now we plot hisograms.  

```python
for i in range(0, 15) :
    if(i!=2) : # because'fnlwgt' is deleted
        plt.figure()
        trainDataFrame[getColumnNameByIndex(i)].hist(legend=True)
        plt.show()
```

## Part 10
### Data normalization

We use `std` method to find standard deviation of a list and `mean` to find average of a list.   
Then we normalize data for all numerical columns in dataframe

```python
trainDataFrame['age'] = ((trainDataFrame['age']-(trainDataFrame['age'].mean())) / (trainDataFrame['age'].std()))
trainDataFrame['education-num'] = ((trainDataFrame['education-num']-(trainDataFrame['education-num'].mean())) / (trainDataFrame['education-num'].std()))
trainDataFrame['capital-gain'] = ((trainDataFrame['capital-gain']-(trainDataFrame['capital-gain'].mean())) / (trainDataFrame['capital-gain'].std()))
trainDataFrame['capital-loss'] = ((trainDataFrame['capital-loss']-(trainDataFrame['capital-loss'].mean())) / (trainDataFrame['capital-loss'].std()))
trainDataFrame['hours-per-week'] = ((trainDataFrame['hours-per-week']-(trainDataFrame['hours-per-week'].mean())) / (trainDataFrame['hours-per-week'].std()))
trainDataFrame['sex'] = ((trainDataFrame['sex']-(trainDataFrame['sex'].mean())) / (trainDataFrame['sex'].std()))

trainDataFrame.head()
```

## PART 11
### PDF Plot

In this part we use `pdf()` method from `norm` field in scipy stats to create PDF of normal distribution of a column and plot it using scatter plot.   
We also use `std` and `mean` to find standard deviation and average of column in order to pass it to `pdf`   
We also get names of dataframe columns with numerical data by using `select_dtypes`   
(Column `sex` is also included)

```python
numericalDataColumns = list((trainDataFrame.select_dtypes(include=numpy.number)).columns)

for colName in numericalDataColumns :
    salaryHIGH50K = trainDataFrame.loc[(trainDataFrame['salary']=='>50K'), colName]
    salaryLOW50K = trainDataFrame.loc[(trainDataFrame['salary']=='<=50K'), colName]
    plt.title(colName)
    plt.scatter(salaryHIGH50K, scp.norm.pdf(salaryHIGH50K, salaryHIGH50K.mean(), salaryHIGH50K.std()), label='>50K')
    plt.scatter(salaryLOW50K, scp.norm.pdf(salaryLOW50K, salaryLOW50K.mean(), salaryLOW50K.std()), label='<=50K')
    plt.legend(['>50K', '<=50K'])
    plt.show()
```

## Part 12
### Reading test data  

In this part we use `read_csv()` method like before to read `test.csv` into a dataframe.  
The `head()` method also makes use sure that data has been successfully read.
We also must normalize data like before in order to make prediction. 

```python
testDataFrame = pd.read_csv('test.csv')
testDataFrame = testDataFrame.replace(['Male', 'Female'] , [0, 1])
testDataFrame['age'] = ((testDataFrame['age']-(testDataFrame['age'].mean())) / (testDataFrame['age'].std()))
testDataFrame['education-num'] = ((testDataFrame['education-num']-(testDataFrame['education-num'].mean())) / (testDataFrame['education-num'].std()))
testDataFrame['capital-gain'] = ((testDataFrame['capital-gain']-(testDataFrame['capital-gain'].mean())) / (testDataFrame['capital-gain'].std()))
testDataFrame['capital-loss'] = ((testDataFrame['capital-loss']-(testDataFrame['capital-loss'].mean())) / (testDataFrame['capital-loss'].std()))
testDataFrame['hours-per-week'] = ((testDataFrame['hours-per-week']-(testDataFrame['hours-per-week'].mean())) / (testDataFrame['hours-per-week'].std()))
testDataFrame['sex'] = ((testDataFrame['sex']-(testDataFrame['sex'].mean())) / (testDataFrame['sex'].std()))

testDataFrame.head(2)
```


### Prediction

In this part we use some of plots created at part 11 to make a prediction of a person salary   

The plots of following dataframe columns are used to predict salary :   

1. `sex`   
If we take a look at this plot created at part 11 we can see that if `sex` value is more than 1 that person probably has a salary less than 50k  `<=50K`   
but else it would probably have a salary more than 50K  `>50K`   
however this prediction might not be completely accurate.    
Therefore we create a function named `getIncomePredictionResult_SEX` to make prediction based on `sex` value.    

```python
def getIncomePredictionResult_SEX(sex_val) :
    if(sex_val < 1) :
        return '>50K'
    else :
        return '<=50K'
```

2. `capital-gain`   
By taking a look at plot distribution (created at part 11) we can see that if `capital-gain` value is between 0.5 and 3.5 that person probably has a salary more than 50k  `>50K`   
but else it would probably have a salary less than 50K  `<=50K`   
The plot distribution is generally good for estimaton (prediction) but like before its not completely accurate.  
Now we create a function named `getIncomePredictionResult_GAIN` to make prediction based on `capital-gain` value.   

```python
def getIncomePredictionResult_GAIN(capitalGain_val) :
    if((capitalGain_val < 3.5) and (capitalGain_val > 0.5)) :
        return '>50K'
    return '<=50K'
```

3. `hours-per-week`   
The plot distribution tells us that if `hours-per-week` value is between 0.4 and 2.4 that person probably has a salary more than 50k  `>50K`   
but else it would probably have a salary less than 50K `<=50K`   
This plot can also be a good estimator.    
Now we create a function named `getIncomePredictionResult_HPW` to make prediction based on `hours-per-week` value.   

```python
def getIncomePredictionResult_HPW(hoursPerWeek_Val) : 
    if((hoursPerWeek_Val > 0.4) and (hoursPerWeek_Val < 2.4)) :
        return '>50K'
    return '<=50K'
```

4. `age`   
The `age` plot distribution also shows that it can be a good estimator.   
As we see if `age` value is between 0 and 2 that person probably has a salary more than 50k  `>50K`   
but else it would probably have a salary less than 50K `<=50K`       
So we create a function named `getIncomePredictionResult_AGE` to make a prediction based on `age` value.   

```python
def getIncomePredictionResult_AGE(age_val) :
    if((age_val > 0) and (age < 2)) :
        return '>50K'
    return '<=50K'
```

5. `education-num`   
The `education-num` can also be a good estimator based on its plot.   
If `age` value is between 0.5 and 2.3 that person probably has a salary more than 50k  `>50K`   
but else it would probably have a salary less than 50K `<=50K`       
So we create a function named `getIncomePredictionResult_EDNUM` to make a prediction based on `age` value. 

```python
def getIncomePredictionResult_EDNUM(educationNum_val) :
    if((educationNum_val > 0.5) and (educationNum_val < 2.3)) :
        return '>50K'
    return '<=50K'
```

Now we use `map` function to map value of each mentioned column to the salary prediction value based on it.    
Then we use `head` function to check our updated dataframe.   

```python
testDataFrame['incomePrediction_sex'] = testDataFrame['sex'].map(getIncomePredictionResult_SEX)
testDataFrame['incomePrediction_GAIN'] = testDataFrame['capital-gain'].map(getIncomePredictionResult_GAIN)
testDataFrame['incomePrediction_HPW'] = testDataFrame['hours-per-week'].map(getIncomePredictionResult_HPW)
testDataFrame['incomePrediction_EDNUM'] = testDataFrame['education-num'].map(getIncomePredictionResult_EDNUM)
testDataFrame['incomePrediction_AGE'] = testDataFrame['age'].map(getIncomePredictionResult_EDNUM)

testDataFrame.head(20)
```
